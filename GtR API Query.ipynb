```python
import sys
import operator
import requests
import json
import twitter
#import pandas as pd

url = 'http://gtr.ukri.org/gtr/api/outcomes?sector=Aerospace'

api = requests.get(url,headers={'Accept': 'application/vnd.rcuk.gtr.json-v7'}) 

#results = api.GetSearch(term='fractals', since=2016-11-21, count=10)

print(api.json())
```

```python
import json

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

jprint(api.json())
```

```python
import pandas as pd

impact_summary = api.json()['impactSummary']
jprint(impact_summary)
```

```python
outcome_id = []

for d in impact_summary:
    time = d['outcomeid']
    outcome_id.append(time)

print(outcome_id)
```

```python
description = []

for d in impact_summary:
    time = d['description']
    description.append(time)

print(description)
```

```python
sector = []

for d in impact_summary:
    time = d['sector']
    sector.append(time)

print(sector)
```

```python
column_names = ['outcome_id', 'description', 'sector']

df = pd.DataFrame(impact_summary) 
print(df)
#useful reference for the above https://thispointer.com/python-pandas-how-to-convert-lists-to-a-dataframe/
```
